```python
from google.colab import files
import pandas as pd
import matplotlib.pyplot as plt

# Upload the file
uploaded = files.upload()

# Read the uploaded file
df = pd.read_csv('penguins.csv')

# Display basic information about the dataset
print(df.info())

# Show the first few rows of the dataset
print(df.head())

# Check for missing values
print(df.isnull().sum())

# Simple analysis: Count the number of penguins by species
species_counts = df['species'].value_counts()
print(species_counts)

# Visualize the distribution of penguin species
plt.figure(figsize=(6, 4))
species_counts.plot(kind='bar', color=['skyblue', 'lightgreen', 'lightcoral'])
plt.title('Distribution of Penguin Species')
plt.xlabel('Species')
plt.ylabel('Number of Penguins')
plt.show()

# Analyze the relationship between flipper length and body mass
plt.figure(figsize=(6, 4))
plt.scatter(df['flipper_length_mm'], df['body_mass_g'], c='orange')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Body Mass (g)')
plt.grid(True)
plt.show()

# Visualize the distribution of bill length by species
plt.figure(figsize=(8, 5))
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.hist(subset['bill_length_mm'], alpha=0.5, label=species)

plt.title('Distribution of Bill Length by Species')
plt.xlabel('Bill Length (mm)')
plt.ylabel('Number of Penguins')
plt.legend()
plt.show()
```
